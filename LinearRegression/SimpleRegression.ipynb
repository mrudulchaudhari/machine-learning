```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("Housing.csv")
df.head()
```

```python
X = df[["area"]] # although there are multiple columns, we are only considering area to learn Simple Linear Regression Model
```

```python
y = df["price"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 42)
```

```python
lr = LinearRegression()
```

```python
X_train
```

```python
y_train
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
print("R2 score", r2_score(y_test, y_pred)) # giving low r2 score again cause we used only area as independent variable
```

```python
print("MSE", mean_squared_error(y_test, y_pred)) # same with this
```
